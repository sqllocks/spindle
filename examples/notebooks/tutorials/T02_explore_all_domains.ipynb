## Import All 12 Domain Classes

### Why this matters

### What to expect

```python
from sqllocks_spindle import (
    RetailDomain,
    HealthcareDomain,
    FinancialDomain,
    SupplyChainDomain,
    IoTDomain,
    HrDomain,
    InsuranceDomain,
    MarketingDomain,
    EducationDomain,
    RealEstateDomain,
    ManufacturingDomain,
    TelecomDomain,
)

print("All 12 domain classes imported successfully!")
```

## List All Domains

### What we're about to do
We'll loop over all 12 domains, instantiate each one, call `get_schema()`, and print a summary table showing:
- **Domain name** — the human-readable label
- **Description** — what business area it models
- **Table count** — how many tables are in the schema
- **Relationship count** — how many foreign-key relationships connect the tables

### Why this matters
This gives you a bird's-eye view of everything Spindle offers out of the box. You can quickly find the domain that matches your project's needs without reading documentation page by page.

### What to expect
A formatted table with one row per domain, showing name, description, tables, and relationships.

```python
domains = [
    RetailDomain(),
    HealthcareDomain(),
    FinancialDomain(),
    SupplyChainDomain(),
    IoTDomain(),
    HrDomain(),
    InsuranceDomain(),
    MarketingDomain(),
    EducationDomain(),
    RealEstateDomain(),
    ManufacturingDomain(),
    TelecomDomain(),
]

print(f"{'Domain':<22} {'Description':<45} {'Tables':>6} {'FKs':>6}")
print("-" * 82)

for domain in domains:
    schema = domain.get_schema()
    print(
        f"{schema.name:<22} "
        f"{schema.description[:43]:<45} "
        f"{len(schema.tables):>6} "
        f"{len(schema.relationships):>6}"
    )

print(f"\nTotal domains: {len(domains)}")
```

## Deep Describe: Retail Domain

### What we're about to do
We'll take the Retail domain and print its complete internal structure:
- Every table name and its column count
- The `generation_order` — the sequence Spindle uses to ensure parent tables are populated before child tables

### Why this matters
Understanding a domain's generation order reveals its dependency graph. Parent tables (like `customers`) are generated first, then child tables (like `orders`) that reference them. This is how Spindle guarantees referential integrity without any post-processing.

### What to expect
A list of tables with column counts, followed by the ordered generation sequence.

```python
retail = RetailDomain()
schema = retail.get_schema()

print(f"Domain: {schema.name}")
print(f"Description: {schema.description}")
print(f"\n=== Tables ({len(schema.tables)}) ===")
for table_name, table_def in schema.tables.items():
    col_count = len(table_def.columns)
    print(f"  {table_name:<30} {col_count:>3} columns")

print(f"\n=== Generation Order ===")
for i, table_name in enumerate(schema.generation_order, 1):
    print(f"  {i}. {table_name}")
```

## Generate and Compare Two Domains Side-by-Side

### What we're about to do
We'll generate both the **Retail** and **Healthcare** domains at the `fabric_demo` scale, then compare:
- Their summaries (table names and row counts)
- How long each took to generate

### Why this matters
Different domains have different complexities — more tables, more relationships, more columns. This comparison helps you understand the performance characteristics and choose appropriate scales for your use case.

### What to expect
Two summary blocks (one per domain) followed by a timing comparison.

```python
import time
from sqllocks_spindle import Spindle

# Generate Retail
t0 = time.perf_counter()
retail_result = Spindle.generate(
    domain=RetailDomain(), scale="fabric_demo", seed=42
)
retail_time = time.perf_counter() - t0

# Generate Healthcare
t0 = time.perf_counter()
healthcare_result = Spindle.generate(
    domain=HealthcareDomain(), scale="fabric_demo", seed=42
)
healthcare_time = time.perf_counter() - t0

print("=== Retail Domain ===")
print(retail_result.summary())

print("\n=== Healthcare Domain ===")
print(healthcare_result.summary())

print("\n=== Timing Comparison ===")
print(f"  Retail:     {retail_time:.2f}s")
print(f"  Healthcare: {healthcare_time:.2f}s")
```
